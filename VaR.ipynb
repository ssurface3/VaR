```python
def calculate_var_hist(data, confidence_level=0.95):
    sorted_data = np.sort(data)
    var_index = int(round((1 - confidence_level) * len(sorted_data),0))
    return sorted_data[var_index]
```

Monte - Carlo

```python
n1= 100
n2 = 1000
n3 = 10000
```

```python
def random_walk(startprice, mu, std,days):
    dt = 1/float(days)
    price = np.zeros(days)
    shock = np.zeros(days)
    price[0] = startprice
    for i in range(1, days):
        shock[i] = np.random.normal(loc=mu * dt, scale=std * np.sqrt(dt))
        price[i] = max(0, price[i-1] + shock[i] * price[i-1])
    return price
```

```python
def simul(mu,days,std, runs,return_var = 0):
    dt = 1/float(days)
    simulations = np.zeros(runs)
    
    for run in range(runs):
        simulations[run] = random_walk(data['adjclose'][1], mu, std,days)[days-1]
    port_return = (simulations - data['adjclose'][1])/data['adjclose'][1]
    q = np.percentile(port_return, 5)
    if return_var == 0:
        plt.hist(port_return, density=True, bins=30, histtype="stepfilled", alpha=0.5)
        plt.figtext(0.15, 0.6, "q(0.95): {:.5}".format(q))
        plt.axvline(x=q, linewidth=1, color="r")
        plt.title("Final return distribution after {} days".format(days), weight="bold")
        plt.show()
    if return_var == 1: 
        return q
```

```python
def BT(var_type,var_value, data,plot_graph = False):
    if plot_graph:
        return 
    data = data['pct_change']
    exceptions_parametric = data < var_value
    num_exceptions = exceptions_parametric.sum()
    print('{} short falls is {}'.format(var_type,num_exceptions))
```

```python

data['pct_change'] = data['adjclose'].pct_change().dropna()
window_size = 100 
confidence_level = 0.95 
data['VaR'] = data['pct_change'].rolling(window=window_size).apply(
    lambda x: np.percentile(x, 100 * (1 - confidence_level)) # parametric way 
)
data['Exception'] = data['pct_change'] < data['VaR']

plt.plot(data.index, data['Daily Return'], label='Daily Returns', color='blue', alpha=0.7)
plt.plot(data.index, data['VaR'], label='Daily VaR (95%)', color='red', linestyle='--')

exception_dates = data[data['Exception']].index
exception_returns = data.loc[exception_dates, 'Daily Return']
plt.scatter(exception_dates, exception_returns, color='orange', label='Exceptions')
plt.show()
```
